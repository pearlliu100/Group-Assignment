## Waffle Chart for Age Group Breakdown

```python
import urllib.request, json 

import pandas as pd

import numpy as np

import geopandas as gpd

import contextily as ctx

import matplotlib.pyplot as plt
```

This is an extension for SB 535 data notebook, we used the number found in the sjoin to create waffle chart for the final analysis of age  demogprahics:

We will be comparing the age breakdown of the Hawthorne/lennox station to the Slauson Station

```python
!pip install pywaffle matplotlib
from pywaffle import Waffle

data = {'Children < 10': 17, '11-64 years': 79, 'Elderly > 64': 4}
fig = plt.figure(
    FigureClass=Waffle, 
    rows=10, 
    values=data, 
    colors=("#FECB52", "#3283F3", "#B6E880"),
    title={'label': 'Age Groups for TOD Station: Hawthorne/Lennox Station', 'loc': 'center'},
    labels=["{0} ({1}%)".format(k, v) for k, v in data.items()],
    legend={'loc': 'lower left', 'bbox_to_anchor': (0, -0.2), 'ncol': len(data), 'framealpha': 0}
)
fig.gca().set_facecolor('#EEEEEE')
fig.set_facecolor('#EEEEEE')
plt.show()
```
