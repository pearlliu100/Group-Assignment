In order to count how many TODs are in each metro line, I use the value.counts function

```python
tod.LINE.value_counts()
tod_line=tod.LINE.value_counts()
tod_line
```

```python
# resetting index,  to make our metro align to the LINE column
tod_line = tod_line.reset_index()
tod_line
```

Let's graph this data!

```python
tod_line.columns=['line','count']
tod_line.plot.bar(x='line',y='count',title='Number of Transit Oriented Districts per metro line')
```

```python
# we only want these three columns, the rest we can drop
tod[['NAME','LINE','geometry']]
```

```python
desired_columns=['NAME','LINE','geometry']
tod_new=tod[desired_columns].copy()
tod_new
```

Let's plot it to see the outcome!

```python
tod_new.plot(
            figsize=(20,20),   
            column = 'LINE',   # column that defines the color of the dots
            legend = True,     # add a legend           
            legend_kwds={
               'loc': 'upper left',
               'bbox_to_anchor':(1,1)
            }                  # this puts the legend to the side
) 
```

```python
fig, ax = plt.subplots(figsize=(10,10))

# create the map plot
tod_new.plot(ax=ax,
         column='LINE',
         legend=True,
         legend_kwds={'loc':'upper left','bbox_to_anchor':(1,1)})

# add a title
ax.set_title('Transit Oriented Developments in Los Angeles County')

# get rid of the axis
ax.axis('off');
```

Add a title for the final touch.

```python
tod_web_mercator = tod_new.to_crs(epsg=3857)
```

```python
fig, ax = plt.subplots(figsize=(10,10))

# create the map plot
tod_web_mercator.plot(ax=ax,
         column='LINE',
         legend=True,
         legend_kwds={'loc':'upper left','bbox_to_anchor':(1,1)})

# add a title
ax.set_title('Transit Oriented Development in Los Angeles County')

# get rid of the axis
ax.axis('off');

ctx.add_basemap(ax)
```

# Folium Interactive Map

I'm hoping to see each TOD and its location, to make the map more interactive, we are going to use folium!

```python
import folium
```

```python
m = folium.Map()
m
```

```python
metro = gpd.read_file('Stations_All_0316.zip')
```

```python
metro.info()
```

```python
desired_columns = ['LINE','LINENUM','STATION','LAT','LONG','geometry']
```

```python
metro[desired_columns].sample(5)
```

```python
metro_trimmed = metro[desired_columns].copy()
metro_trimmed
```

Now I want to subset the data using loc function to pick out the TOD station.

```python
metro_tod=metro_trimmed.loc[(metro_trimmed['STATION'] == "Florence Station") | 
                            (metro_trimmed['STATION'] == "Slauson Station") |
                           (metro_trimmed['STATION'] == "Firestone Station")|
                           (metro_trimmed['STATION'] == "Hawthorne / Lennox Station")|
                           (metro_trimmed['STATION'] == "Aviation / LAX Station")|
                           (metro_trimmed['STATION'] == "Sierra Madre Villa Station")|
                            (metro_trimmed['STATION'] == "Del Amo Station")|
                            (metro_trimmed['STATION'] == "East LA Civic Center Station")]
metro_tod
```

```python
for index, row in metro_tod.iterrows():
    print(row.STATION, row.LAT, row.LONG)
```

```python
for index, row in metro_tod.iterrows():
    # add folium marker code
    folium.Marker([row.LAT, row.LONG], popup=row.STATION, tooltip=row.STATION).add_to(m)
m
```

## Color code markers

```python
metro_tod['color'] = ''
```

```python
metro_tod.head()
```

```python
metro_tod.LINE.unique()
```

```python
metro_tod.loc[metro_tod['LINE'] == 'Blue']
```

```python
metro_tod.loc[metro_tod['LINE'] == 'Blue', 'color'] = 'blue'
metro_tod.loc[metro_tod['LINE'] == 'Green', 'color'] = 'green'
metro_tod.loc[metro_tod['LINE'] == 'Gold', 'color'] = 'orange'
```

```python
metro_tod.sample()
```

```python
# add the stations with color icons
for index, row in metro_tod.iterrows():
    tooltip_text = row.LINE + ' Line: ' + row.STATION
    folium.Marker(
        [row.LAT,row.LONG], 
        popup=row.STATION, 
        tooltip=tooltip_text,
        icon=folium.Icon(color=row.color)
    ).add_to(m)

# show the map
m
```

Now we have a color coded interactive map to show all the TOD stations across LA County!
