# Overlaying and Colorcoding

Using the example of traffic percentile more than 50% with TOD Stations

```python
fig, ax = plt.subplots(figsize=(12, 12))

# set a title
ax.set_title('Traffic Percentile more than 50% with TOD Stations',fontsize=24,pad=20)


# create the map plot

colors = {'BLUE': "#9BE0E0", 'GOLD': "#FAE90A", 'GREEN': "#42FF75",'SILVER': "#C5DBD8"}
h = [colors.get(x,)for x in tod_web_mercator.LINE]

plt.figure() 

tod_web_mercator.plot(
        ax=ax,
        color=h)

#manually add handles and labels
blue_patch = mpatches.Patch(color='#9BE0E0', label='Blue Line')
gold_patch =mpatches.Patch(color='#FAE90A',label='Gold Line')
green_patch =mpatches.Patch(color='#42FF75',label='Green Line')
silver_patch =mpatches.Patch(color='#C5DBD8',label='Silver Line')
ax.legend(handles=[blue_patch, gold_patch, green_patch, silver_patch ])
tod_web_mercator.boundary.plot(ax=ax,color='black',zorder=20,lw=0.75)


# get rid of the axis
ax.axis('off');


# add the layer with ax=ax in the argument 
sb535[sb535['Traffic_Pctl'] > 50].plot(ax=ax, 
                                color="red",
                                alpha=0.2)
# turn the axis off
ax.axis('off')


# add a basemap
ctx.add_basemap(ax,source=ctx.providers.Stamen.TonerLite)
```

Making the chlorpleth map a function!

```python
def livable_tod_metrics(sb535_metric):
    fig, ax = plt.subplots(figsize=(12, 12))
    colors = {'BLUE': "#9BE0E0", 'GOLD': "#FAE90A", 'GREEN': "#42FF75",'SILVER': "#C5DBD8"}
    h = [colors.get(x,)for x in tod_web_mercator.LINE]
    plt.figure() 
    tod_web_mercator.plot(
        ax=ax,
        color=h)
    
    #manually add handles and labels
    blue_patch = mpatches.Patch(color='#9BE0E0', label='Blue Line')
    gold_patch =mpatches.Patch(color='#FAE90A',label='Gold Line')
    green_patch =mpatches.Patch(color='#42FF75',label='Green Line')
    silver_patch =mpatches.Patch(color='#C5DBD8',label='Silver Line')
    ax.legend(handles=[blue_patch, gold_patch, green_patch, silver_patch ])
    tod_web_mercator.boundary.plot(ax=ax,color='black',zorder=20,lw=0.75)
    
    # turn the axis off
    ax.axis('off')

    sb535[sb535[sb535_metric] > 80].plot(ax=ax, 
                                color="red",
                                alpha=0.2)
    ax.axis('off')
    ax.set_title(sb535_metric +' more than 80% with TOD Stations',fontsize=24,pad=20)
    
    # add a basemap
    ctx.add_basemap(ax,source=ctx.providers.Stamen.TonerLite)
```

making sure that our data use the same axis

```python
tod_web_mercator.crs
```

```python
sb535.crs
```

```python
%%time
# run the function once
livable_tod_metrics(sb535_metric = 'Linguistic_Isolation_Pctl')
```

```python
sb535[sb535['Linguistic_Isolation_Pctl'] > 80].plot()
```

```python
metrics_list = ['Pollution_Burden_Pctl','Traffic_Pctl', 'Housing_Burden_Pctl']
metrics_list
```

```python
%%time
# run our function for every percentile in our in our list, create a for loop to go through the function
for sb535_metric in metrics_list:
    livable_tod_metrics(sb535_metric)
```

# Spatial Join

spatial joining the TOD stations to the SB 535 data

```python
metro = gpd.read_file('Stations_All_Data.zip') 
```

```python
desired_columns = ['LINE','LINENUM','STATION','LAT','LONG','geometry']
```

```python
metro_trimmed = metro[desired_columns].copy()
```

```python
metro_tod=metro_trimmed.loc[(metro_trimmed['STATION'] == "Florence Station") | 
                            (metro_trimmed['STATION'] == "Slauson Station") |
                           (metro_trimmed['STATION'] == "Firestone Station")|
                           (metro_trimmed['STATION'] == "Hawthorne / Lennox Station")|
                           (metro_trimmed['STATION'] == "Aviation / LAX Station")|
                           (metro_trimmed['STATION'] == "Sierra Madre Villa Station")|
                            (metro_trimmed['STATION'] == "Del Amo Station")|
                            (metro_trimmed['STATION'] == "East LA Civic Center Station")]
```

```python
metro_tod
```

```python
metro_tod.plot()
```

```python
sb535 = sb535.to_crs(epsg=4326)
```

we see that the it was succesive!

## Ploty Chart

here was begining the process for the plotly charts

```python
import plotly.express as px 
```

```python
sjoined_metrics
```

```python
# ok, group by date and charge, and let's get a count for each
a_type=sjoined_metrics.groupby(['STATION','Traffic_Pctl']).size().reset_index(name='count')
```

```python
# make a stacked bar chart
fig = px.bar(a_type,
           x='STATION',
           y='Traffic_Pctl',
           color='Traffic_Pctl', # this creates the "stack",
           labels={'STATION':'station','count':'Percentile','Traffic_Pctl':'traffic percentile'},
           title = 'Traffic Percentile'
      )
fig.show()
```
