```python
tod_new.loc[tod_new['LINE'] == 'BLUE', 'color'] = 'blue'
tod_new.loc[tod_new['LINE'] == 'GREEN', 'color'] = 'green'
tod_new.loc[tod_new['LINE'] == 'GOLD', 'color'] = 'orange'
tod_new.loc[tod_new['LINE'] == 'SILVER', 'color'] = 'gray'
```

```python
TODPalette = {'BLUE': 'blue',
               'GREEN': 'green',
               'GOLD': 'orange',
               'SILVER': 'gray'}
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

# Loop through each attribute type and plot it using the colors assigned in the dictionary
for ctype, data in tod_new.groupby('LINE'):
    
    # Define the color for each group using the dictionary
    color = TODPalette[ctype]
    
    # Plot each group using the color defined above
    data.plot(color=color,
              ax=ax,
              label=ctype
             )

    ax.legend(fontsize=15,
          frameon=False,
          loc=('lower right'),
          title="LEGEND")
    
ax.set(title='Transit Oriented Development in Los Angeles County')
        
             
             


# get rid of the axis
ax.axis('off');


plt.show()
ctx.add_basemap(ax)
```

```python
def livable_tod_metrics(sb535_metric = 'Traffic_Ptcl'):
    
    gdf_web_mercator = sb535.to_crs(epsg=3857)                           
   
    fig, ax = plt.subplots(figsize=(12, 12))

    # add the layer with ax=ax in the argument 
    gdf_web_mercator[gdf_web_mercator[sb535_metric] > 80].plot(
                                                               ax=ax, 
                                                               color="orange",
                                                               alpha=0.8
                                                                )

    # turn the axis off
    ax.axis('off')

    # set a title
    ax.set_title('Traffic Percentile more than 80% with TOD Stations',fontsize=24,pad=20)


    # create the map plot
    tod_web_mercator.plot(ax=ax,
         column='LINE',
         legend=True,
         legend_kwds={'loc':'upper left','bbox_to_anchor':(1,1)})

    # get rid of the axis
    ax.axis('off');

    ctx.add_basemap(ax)


    # add a basemap
    ctx.add_basemap(ax,source=ctx.providers.Stamen.TonerLite)
```

```python
%%time
# run the function once
livable_tod_metrics(sb535_metric = 'Traffic_Ptcl')
```
