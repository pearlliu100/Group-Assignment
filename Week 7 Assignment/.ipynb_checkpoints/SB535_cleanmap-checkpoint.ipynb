```python
TODPalette = {'BLUE': 'blue',
               'GREEN': 'green',
               'GOLD': 'orange',
               'SILVER': 'gray'}
```

```python
fig, ax = plt.subplots(figsize=(15, 15))

# Loop through each attribute type and plot it using the colors assigned in the dictionary
for ctype, data in tod_new.groupby('LINE'):
    
    # Define the color for each group using the dictionary
    color = TODPalette[ctype]
    
    # Plot each group using the color defined above
    data.plot(color=color,
              ax=ax,
              label=ctype
             )

    ax.legend(fontsize=10,
          frameon=False,
          loc=('lower right'),
          title="LEGEND")
    
ax.set(title='Transit Oriented Development in Los Angeles County')
        
             

# get rid of the axis
ax.axis('off');


plt.show()
ctx.add_basemap(ax)
```

## Function and for loop: TOD Layer and SB 535 Data

```python
def livable_tod_metrics(sb535_metric):
    
    gdf_web_mercator = sb535.to_crs(epsg=3857)                           
   
    fig, ax = plt.subplots(figsize=(12, 12))

    # add the layer with ax=ax in the argument 
    gdf_web_mercator[gdf_web_mercator[sb535_metric] > 70].plot(
                                                               ax=ax, 
                                                               color="orange",
                                                               alpha=0.8
                                                                )

    # turn the axis off
    ax.axis('off')

    # set a title
    ax.set_title(sb535_metric + ' Percentile more than 70% near Transit Oriented Districts',fontsize=24,pad=20)


    # create the map plot
    tod_web_mercator.plot(ax=ax,
         column='LINE',
         legend=True,
         legend_kwds={'loc':'upper left','bbox_to_anchor':(1,1)})

    # get rid of the axis
    ax.axis('off');

    ctx.add_basemap(ax)


    # add a basemap
    ctx.add_basemap(ax,source=ctx.providers.Stamen.TonerLite)
```

Here is our test drive with one of the metrics within the function, let's try Lingustic Isolation Percentile

```python
%%time
# run the function once
livable_tod_metrics(sb535_metric = 'Linguistic_Isolation_Pctl')
```

Okay! It works! Now let's make array for a for loop!

```python
metrics_list = ['Pollution_Burden_Pctl','Traffic_Pctl', 'Housing_Burden_Pctl','Unemployment_Pctl']
metrics_list                
```

So these percentile metrics were the ones that piqued our interest. We might cut some down later, but these parrallel to the livability aspect.

Here's our for loop:

```python
%%time
# run our function for every address in our list
for sb535_metric in metrics_list:
    livable_tod_metrics(sb535_metric)
```
