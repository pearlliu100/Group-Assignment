```python
gmap['Percentage Male'] = gmap['Male:']/gmap['Total']*100
gmap['Percent Male 50 to 54 years'] = gmap['50 to 54 years']/gmap['Total']*100
gmap['Percent Male 55 to 59 years'] = gmap['55 to 59 years']/gmap['Total']*100
gmap['Percent Male 60 and 61 years'] = gmap['60 and 61 years']/gmap['Total']*100
gmap['Percent Male 62 to 64 years'] = gmap['62 to 64 years']/gmap['Total']*100
gmap['Percent Male 65 and 66 years'] = gmap['65 and 66 years']/gmap['Total']*100
gmap['Percent Male 67 to 69 years'] = gmap['67 to 69 years']/gmap['Total']*100
gmap['Percent Male 70 to 74 years'] = gmap['70 to 74 years']/gmap['Total']*100
gmap['Percent Male 85 years and over'] = gmap['85 years and over']/gmap['Total']*100
gmap['Percentage Female'] = gmap['Female:']/gmap['Total']*100
gmap['Percent Female 50 to 54 years'] = gmap['F:50 to 54 years']/gmap['Total']*100
gmap['Percent Female 85 years and over'] = gmap['F:85 years and over']/gmap['Total']*100
                                              
                                             
                                                                      
                                                                
    

                                        
```

assigned percentages to the different age groups by sex and then printed it 

```python
gmap.plot(figsize=(12,12),
          column= 'Percent Male 50 to 54 years',
          scheme ='user_defined',
          classification_kwds={'bins':[ 0, 5, 15, 20]}, 
          edgecolor = 'white',
          cmap ='YlOrRd',
         legend= True)
```

Map of Percentage Male by census tracts

```python
gmap.plot(figsize=(12,12),
          column= 'Percent Female 50 to 54 years',
          legend = True,
          scheme = 'equal_interval',
          cmap = 'Blues')
```

Map of Percentage Female by equal interval

```python
gmap_web_mercator = gmap.to_crs(epsg=3857)
```

call mercator to help subplot

```python
fig, ax = plt.subplots(figsize=(15,15))

gmap_web_mercator[gmap_web_mercator['Percentage Male'] > 50].plot(
                                                    ax = ax,
                                                    alpha = 0.5,
                                                    color = "red")
ax.axis('off')
ax.set_title('Census Tracts with more than 50% Male population', fontsize= 24, pad =20)
ctx.add_basemap(ax)
```

create multiple layers to higlight Census tract where the female population is more than 50%

```python
fig, ax = plt.subplots(figsize=(15,15))

gmap_web_mercator[gmap_web_mercator['Percentage Female'] > 50].plot(
                                                    ax = ax,
                                                    alpha = 0.5,
                                                    color = "blue")
ax.axis('off')
ax.set_title('Census Tracts with more than 50% Female population', fontsize= 24, pad =20)
ctx.add_basemap(ax)
```
